```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel("C:/Users/koushik.n/Downloads/Diwali Sales Data Rev.xlsx")
df
```

```python
df.shape
```

```python
df.info()
```

Data cleaning

```python
df.isna().sum()
```

```python
df.drop(columns=['Status','unnamed1'], inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna(how='all')
df
```

```python
#dropping the rows with na values
```

```python
df = df.dropna()
df
```

```python
df.info()
```

```python
# replacing the marital status column with meaningful values. Replacing 0 with 'Unmarried' and 1 with 'Married'
```

```python
df['Marital_Status'].replace({0 : 'Unmarried', 1 : 'Married'}, inplace=True)
df.head(10)
```

```python
# replacing the Gender column with meaningful values. Replacing M with 'Male' and F with 'Female'

df['Gender'].replace({'F':'Female', 'M':'Male'}, inplace=True)
df
```

```python
df['Amount'] = df['Amount'].astype('int32')
df.head()
```

# Exploratory Data Analysis

# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)


```python
df.columns
```

```python


sales_gender = df.groupby('Gender')['Amount'].sum()
sales_gender
```

```python
# plotting a bar chart for gender vs total amount
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gender)
```

```python
#adding an index column, otherwise getting this error
```

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum()
sales_gender
```

```python
sns.barplot(x='Gender', y='Amount', data = sales_gender)
```

# In terms of Order count and Order value female customers have purchased more than male customers

```python
#now trying to plot the data by the age group, in terms of order count as well as order value

ax = sns.countplot(x= 'Age Group' , hue='Gender', data=df)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sales_age
```

```python
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

# With the help of the above charts we can see that most of the buyers belong to the 26-35 age group, in terms of both order count and order value

```python
df.columns
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data= sales_state)
```

```python
sales_state_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sales_state_amount
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', data= sales_state_amount)
```

# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
